### Praktikum 1 : Tipe Data Dasar

```python
angka_int = 10
print("angka_int =", angka_int, "->", type(angka_int))

angka_float = 3.14
print("angka_float =", angka_float, "->", type(angka_float))

angka_complex = 2 + 3j
print("angka_complex =", angka_complex, "->", type(angka_complex))

is_active = True
print("is_active =", is_active, "->", type(is_active))

teks = "Hello, Python!"
print("teks =", teks, "->", type(teks))

daftar_angka = [1, 2, 3, 4, 5]
print("daftar_angka =", daftar_angka, "->", type(daftar_angka))

koordinat = (10, 20)
print("koordinat =", koordinat, "->", type(koordinat))

data_mahasiswa = {
    "nama": "Andi",
    "nim": "A11.2022.12345",
    "jurusan": "Teknik Informatika"
}
print("data_mahasiswa =", data_mahasiswa, "->", type(data_mahasiswa))

himpunan_angka = {1, 2, 3, 2, 1}
print("himpunan_angka =", himpunan_angka, "->", type(himpunan_angka))

nilai_str = "100"
nilai_int = int(nilai_str)
print("\nnilai_int =", nilai_int, "->", type(nilai_int))
```

### Praktikum 2 : Variabel dan Operasi Dasar

```python
nama = "Budi"
umur = 20
tinggi = 170.5
is_student = True

print("Nama =", nama)
print("Umur =", umur)
print("Tinggi =", tinggi, "cm")
print("Mahasiswa =", is_student)

a = 10
b = 3

penjumlahan = a + b
pengurangan = a - b
perkalian = a * b
pembagian = a / b
pembagian_bulat = a // b
modulus = a % b
pangkat = a ** b

print("\nOPERASI ARITMETIKA")
print("Penjumlahan =", penjumlahan)
print("Pangkat =", pangkat)

print("\nOPERASI PERBANDINGAN")
print("a > b =", a > b)
print("a == b =", a == b)

x = True
y = False
print("\nOPERASI LOGIKA")
print("x and y =", x and y)
print("not x =", not x)
```

### Program 2 : Cek Bilangan Genap, Ganjil, atau Prima

```python
bilangan = int(input("Masukkan sebuah bilangan bulat: "))

if bilangan % 2 == 0:
    status_kali = "Genap"
else:
    status_kali = "Ganjil"

is_prima = True
if bilangan < 2:
    is_prima = False
else:
    for i in range(2, int(bilangan**0.5) + 1):
        if bilangan % i == 0:
            is_prima = False
            break

status_prima = "Prima" if is_prima else "Bukan Prima"

print(f"Bilangan {bilangan} adalah bilangan {status_kali} dan {status_prima}.")
```
